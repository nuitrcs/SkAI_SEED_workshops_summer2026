# Challenge: How far away are the nearest stars?

The file `stellar_data.txt` is a real catalog of the Sun's 26 nearest stellar systems. In this challenge, you will read in these data, calculate the distance to each star, write your results to a new file, and check them.

**1. Read the file.** It's a fixed-width astronomy table, so watch for a few things:
- The columns are separated by spaces (which string method splits on whitespace?).
- The first three lines are header text, not data. 
- The columns you need are: `Name`, `V` (apparent magnitude), and `M_V` (absolute magnitude).  `V` and `M_V` should be floats.
- Not every line has the same number of columns with data.  
- Consider using `try`/`except` and/or conditionals to catch potential errors as you read in the file and store the data.

**2. Calculate each distance** with the *distance modulus*. For apparent magnitude $V$ and absolute magnitude $M_V$, the distance $d$ in parsecs is:

$$d = 10^{\,(V - M_V + 5)\,/\,5}$$

Write this as a function and round the answer to 2 decimal places.

**3. Write to file.** Write each star's name and distance to a new csv file named `star_distances.csv`.

**4. Check your work.** Read your `star_distances.csv` file back and print the contents. Your distances should match the catalog's own `Distance` e.g. Proxima Centauri should be **1.29 pc** and Sirius A **2.64 pc**.  Check this explicitly with code.

**A plan (one of many possibilities):**
1. Open `stellar_data.txt` and read all its lines.
2. Loop over the lines and `split()` each one on whitespace.
3. `try` to read the columns of interest and store as the magnitudes as floats; on failure, skip the line (it's a header).
4. Check how many columns are available in each row and print a warning if less than expected.
5. Store each star's name, apparent magniude and absolute magnitude in list.
6. Write a `distance_pc(V, MV)` function using the formula; loop to get every distance.
7. Open a new file in write mode and write `name,distance` for each star.
8. Read the new file back in and print the output to check
9. Compare my distances against the catalog's own distance column.

```python
# 1. Open `stellar_data.txt` and read all its lines.
filename = "stellar_data.txt"
with open(filename, "r") as f:
    lines = f.readlines()

# 2. Loop over the lines and `split()` each one on whitespace.
stars = []
for line in lines:
    parts = line.split()
    # 3. `try` to read the columns of interest and store as the magnitudes as floats; on failure, skip the line (it's a header).
    try:
        name = parts[0]
        V = float(parts[3])
        MV = float(parts[4])
        distance_check = float(parts[7])
    except ValueError:
        print(f"ERROR: cannot convert {parts[3]}, {parts[4]} or {parts[7]} to floats.  Suspected header line.  Skipping...")
        continue
    except IndexError:
        print(f"ERROR: line contains too few columns ({len(parts)}).  Skipping ...")
        continue

    # 4. Check how many columns are available in each row and print a warning if less than expected.
    if (len(parts) < 9):
        print(f"WARNING: row for star {name} contains less columns ({len(parts)}) than expected.  ")

    # 5. Store each star's name, apparent magniude and absolute magnitude in list.
    stars.append([name, V, MV, distance_check])

print(len(stars), "stars read in")
```

```python
# quick check to make sure I read in the data as expected
stars
```

```python
# 6. Write a `distance_pc(V, MV)` function using the formula; loop to get every distance.
def distance_pc(V, MV):
    """Distance in parsecs from apparent magnitude V and absolute magnitude MV."""
    return round(10 ** ((V - MV + 5) / 5), 2)

# 7. Open a new file in write mode and write `name,distance` for each star.
with open("star_distances.csv", "w") as f:
    f.write("name,distance_pc\n")
    for star_values in stars:
        # calculate the distance using my function
        d = distance_pc(star_values[1], star_values[2])

        # I will save the calculated distance here so that I can compare to the file's value later
        star_values.append(d) 

        # write to the file
        f.write(star_values[0] + "," + str(d) + "\n")
```

```python
# 8. Read the new file back in and print the output to check
with open("star_distances.csv", "r") as f:
    print(f.read())
```

```python
# 9. Compare my distances against the catalog's own distance column.
# (I'm using some more advanced f string formatting to align the values for easier viewing)
print(f"{'Name':<34} {'file_distance':<14} {'my_distance':<14} {'difference':<14}")
print("-" * 75) # Separator line

for star_values in stars:
    print(f"{star_values[0]:<34} {star_values[3]:<14} {star_values[4]:<14} {round(star_values[4] - star_values[3], 2):<14}")
```
