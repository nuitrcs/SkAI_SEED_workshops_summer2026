Python has a basic way to open files, but I'm going to teach you the better way. The way I teach you is the way all Python coders open files. You may someday encounter a logic situation where you need to use the old way to open a file, so I'll show the syntax to you briefly.

```{python}
f = open(filename, "r")
#do something with the file
f.close()
```

This leaves the file needlessly open until you close it, which takes up memory. It also leaves you open to potentially forgetting to close the file.

Files tend to take up more memory inside Python than other Python objects like strings, lists, and dictionaries.

### `with`/`as` statement: The better way to open files 

Here is the syntax to read a file. (This code isn't ready to run, it's just to look at to see the syntax.)

```{python}
with open(filename, "r") as f:
    # save file as some other object
    # or save part of a file
```

The first line is the `with`/`as` statement. The `f` is a temporary variable that will store the **file object**. Just like in a for loop, you can use anything for the temporary variable, but `f` is commonly used.

**Inside** the `with`/`as` statement, you want to save the file as a different object type - something that doesn't take up as much memory as a file.

The file will automatically close when we **exit** the with/as statement (meaning when we exit the indentation).

### The `open` function

The `open()` function takes two arguments: the filename and the mode.

Mode options:
- "r"  read
- <span style="color:red">"w"  write (wipes the file clean if it already exists)
- "a"  append (add to the end of whatever is already in the file)


### Filenames

If you are accessing a file in your current working directory, you can just include the filename, but if the file is in a different directory, you must include either the relative or absolute path.

Let's try opening the file "wht20250414.txt" and printing it to see what it looks like. We will use the read mode:

````python
with open(wht_2025_Apr_14_obslog_filename, "r") as f:
    print(f)
````

### 6. Turning a file into a dictionary

Store the filename that we will be working with in the next few examples:

````python
comets_in = "NEOcomets.csv"
````

We'll work through this example together. Let's open this file and see what it looks like:

````python
with open(comets_in, "r") as f:
    print(f.read())
````

**<span style="color:crimson">LOGIC** **Our end goal is to have a dictionary with the spkid name as the key and a list of the remaining itesm as the values.**

*Note*: Later we will teach you other ways to read in `.csv` files (and other file types with columnated data) that may help accelerate your research workflows.

First let's store the file contents as a list, but here we want to leave out the first line of headers. When we call `f.readlines()` it turns the file into a list. We can index a list, so let's take all the lines except the first one:

````python
with open(comets_in, "r") as f:
    comets = f.readlines()[1:]
````

Now let's view our new list:

````python
for line in comets:
    print(line)
````

We can see that there are new line characters at the end of each line (because it is printing extra empty lines between the lines). Let's make a note of that.

We can apply what we know about lists and strings to make a list of what we need to code:
- make an empty dictionary
- loop through the comets list
- remove the new line characters from the end
- split the line on the commas
- separate the first item to be the key
- store the rest of the items as a list
- assign the key:value pairs to our dictionary

I've put **comments** in the code to remind us of what we need to do. Comments start with a `#` symbol and are ignored by the computer.

````python
comets_dict = {}             # make an empty dictionary
for line in comets:          # loop through the gradebook list
    line = line.rstrip("\n") # remove the new line characters from the end
    line = line.split(",")   # split the line on the commas
    key = line[0]            # separate the first item to be the key
    value = line[1:]         # store the rest of the items as a list
    comets_dict[key] = value # assign the key:value pairs to our dictionary
````

````python
print(comets_dict)
````

**Question:** We wrote out all the steps and wrote the code with one line per step. This code could be condensed into fewer lines or left how it is - as explicit as possible. **Can you think of ways that the code could be condensed to fewer lines?**

````python
comets_dict = {}
for line in comets:
    line = line.rstrip("\n").split(",")
    comets_dict[line[0]] = line[1:]
print(comets_dict)
````

### 7. Reading files line by line - demo

Sometimes you might be working with a very large file, with millions of lines, and you don't want to read it all into memory as a string or list.

There is a file method, `readline()`, that reads in only one line at a time. **I don't expect you to practice this method here, but I will give an example, so that you know it exists if you ever need to look it up.**

Let's imagine that there are millions of NEO comets (!!) and our NEOcomets.csv file is millions of lines long. We can use readline to loop through the file and only store the comets that we need for this notebook or script. This doesn't work the same way as `readlines()` because `readlines()` is a list and `readline()` is the string of only the first line. We need to use a while loop, which is something we aren't learning this week:

````python
comets_file = "NEOcomets.csv"
````

````python
panstarrs_NEO_comets = []
with open(comets_file, "r") as f:
    line = f.readline()
    while line:
        if "PANSTARRS" in line:
            panstarrs_NEO_comets.append(line.rstrip("\n"))
        line = f.readline()
````

````python
for x in panstarrs_NEO_comets:
    print(x)
````
