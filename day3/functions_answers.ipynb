# Defining your own functions


You already know how to **call** a function.

````python
len("How long is this string?")
````

````python
round(6479.382029, -2)
````

You can also write your own custom functions. Why would you want to do that?
- If you find yourself using the same code repeatedly, so that you don't have to write it over and over
- If you want to break your code up into chunks to make it much more readable

To create our own function, we create a **function definition**.

The function definition starts with a **`def` statement**.

The next line (inside the indentation) should be a short **docstring** that says what your function does. This is just good practice. A docstring is like a comment - it is ignored by the computer. A docstring is contained in **triple quotation marks** like `''' Here is my docstring. '''`

Next (still inside the `def` statement), you write the code for what the function does.

Here is the syntax. (This code won't work, it's just to look at.)

```python
def function_name(arguments, if_needed):
   '''a useful comment'''
   # do something or create a new object
   # return something (optional)
```

## 2. Writing a function with one argument

We can add an argument. Whatever you call the arguments in your function definition must match exactly to how they are used inside the function definition, just like we saw with for loops and with/as statements:

````python
def hello_you(name):
    '''Prints Hello You! replacing You with whatever string you give it.'''
    print("Hello " + name + "!")
````

Now we can pass it any string as an argument:

````python
hello_you("Eeyore")
````

`name` is only used within the function - it doesn't exist outside the function. We say it is out of **scope**.

````python
name
````

## 3. Writing a function that returns a value

Let's write our own function to find the area of a rectangle.

The arguments our function will need are length and width. 

````python
def area(length, width):
    '''This function takes a length and width of a rectangle and returns the area.'''
    answer = length * width
````

````python
area(10, 12)
````

````python
print(answer)
````

We created `answer` inside our function definition, but it doesn't exist outside that definition. We need to include a **return statement** if we want our function to return the value of an object created inside the function.

````python
def area(length, width):
    '''This function takes a length and width of a rectangle and returns the area.'''
    answer = length * width
    return answer
````

````python
area(10, 12)
````

Like any function, we can assign the output of a custom function to a variable. Let's say my kitchen is 10 feet long and 12 feet wide:

````python
kitchen_area = area(10, 12)
````

````python
print(kitchen_area)
````

Also like other functions, we can pass variables to the function as our arguments:

````python
kitchen_l = 10
kitchen_w = 12
````

````python
kitchen = area(kitchen_l, kitchen_w)
print(kitchen)
````

## Exercise: Function practice 3

Write a function that applies **Kepler's Third Law** :  given a list of orbital periods (T) the function should calculate and return the semi-major axis (a), or vice versa if you prefer.  For reference, here are the equations you need: 

$$T^2 = \frac{4\pi^2}{G(M + m)} a^3$$

Solving for $a$:

$$a = \left(\frac{G(M + m) T^2}{4\pi^2}\right)^{1/3}$$


`G` is the Gravitational constant. 

Round your answer to the nearest 2 decimal places.

Test your function using the mass of the Sun and Earth, and the Earth's orbital period (i.e., 1 year).

*Bonus*: 
- Write your function (or functions) so that it can work for both a single period value and a list of period values.
- Add a unit converter (like we did above) and a new input argument to choose the output unit.

````python
# Hint: you can get pi from the math library
import math
print(math.pi)
````

````python
def kepler_period_to_sma(period, m1, m2):
    ''' Converts an orbital period to a semi-major axis.

    This function uses Kepler's Third law to convert an orbital period to 
    a semi-major axis, given masses m1 and m2
    
    Args:
        period: The orbital period in days
        m1: mass of the primary body in solar masses
        m2: mass of the secondary body in solar masses

    Returns:
        The semi-major axis in AU, rounded to two decimal places

    '''
    G = 2.959e-4  # AU^3 / (M_sun * day^2)
    sma = (G * (m1 + m2) * period**2 / (4*math.pi**2))**(1/3)

    return round(sma,2)
````

````python
# test for Sun and Earth
kepler_period_to_sma(365.25, 1, 3e-6)
````

````python
# Bonus to take a list of period values and add a unit converter
AU_dict = {"parsec": 4.8481e-6, "light_year": 1.5813e-5, "km": 1.496e+8, "mile":  9.296e+7}
def kepler_period_to_sma(period, m1, m2, converter = None):
    '''rewrite our initial function to add the unit converter'''

    G = 2.959e-4  # AU^3 / (M_sun * day^2)
    sma = (G * (m1 + m2) * period**2 / (4*math.pi**2))**(1/3)

    if converter is not None:
        sma = sma * AU_dict[converter]

    return sma

def periods_to_sma(periods, m1s, m2s, converter = None):
    '''use this as a wrapper to allow for a list of periods and masses, all returning the same unit'''
    return [round(kepler_period_to_sma(p,m1,m2, converter=converter),2) for (p,m1,m2) in zip(periods, m1s, m2s)]
````

````python
# for all planets
planet_names   = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]
planet_masses  = [1.652e-7, 2.447e-6, 3.003e-6, 3.213e-7, 9.543e-4, 2.857e-4, 4.366e-5, 5.151e-5]  # M_sun
planet_periods = [87.97, 224.70, 365.25, 686.97, 4332.59, 10759.22, 30688.50, 60195.00]  # days


periods_to_sma(planet_periods, [1]*len(planet_periods), planet_masses)
#periods_to_sma(planet_periods, [1]*len(planet_periods), planet_masses, 'km')
````

````python
# Note that the masses of our planets are so small compared to the Sun that you
# could neglect the planet masses and still get the right answers 
# (to w/in 2 decimal places)

periods_to_sma(planet_periods, [1]*len(planet_periods), [0]*len(planet_periods))
````

2. Write a function that calculates and returns the escape velocity (`v_esc`) given inputs of the object's mass ($M$) and radius ($R$).  The relevant formula is:

    $$v_{esc} = \sqrt{\frac{2GM}{R}}$$

    (Where $G$ is the gravitational constant.)  Test that your function returns an escape velocity for Earth of 11.186 km/s.

````python
def vesc(m, r):
    '''calculate the escape velocity for a given mass and radius
    
    Uses the equation:
    v_esc = sqrt(2GM / R)
    
    Args:
        m: mass of the object in Earth masses
        r: radius of the object in Earth radii

    Returns:
        The escape velocity in km/s
    '''

    G = 1.541e-6  # R_earth^3 / (M_earth * s^2)
    REarth =  6371 # Earth radius in km

    return (2 * G * m / r)**0.5 * REarth

vesc(1,1)
````

3. Add to your function to allow the user to define the input and output units within a set of available options that you define (e.g., you could allow inputs of Solar units or Earth units, and choose outputs of km/s or mph). 

    Write down the steps you will need, and then try to write the code.  **It is OK if your code doesn't work on the first attempt.** Flex your debugging muscles! And ask for help if you need it.

    *Bonus:* add Error handling that checks if the requested units are valid.

````python
def vesc(m, r, in_units = "Earth", out_units = "kps"):
    '''calculate the escape velocity for a given mass and radius
    
    Uses the equation:
    v_esc = sqrt(2GM / R)
    
    Args:
        m: mass of the object in "in_units"
        r: radius of the object in "in_units"
        in_units : units of the input mass and radii, either "Earth" or "Sun"
        out_units : desired output units, either "kps" or "mph"

    Returns:
        The escape velocity in km/s
    '''

    # define look-up tables as dicts with tuple keys for ease of use
    G_dict = {
        ("Earth", "kps"): 1.541e-6,    # R_earth^3 / (M_earth * s^2)
        ("Earth", "mph"): 19.97136,    # R_earth^3 / (M_earth * hr^2)
        ("Sun",   "kps"): 3.949e-7,    # R_sun^3   / (M_sun * s^2)
        ("Sun",   "mph"): 5.117904,    # R_sun^3   / (M_sun * hr^2)
    }

    R_dict = {
        ("Earth", "kps"): 6371,        # km
        ("Earth", "mph"): 3959,        # miles
        ("Sun",   "kps"): 695700,      # km
        ("Sun",   "mph"): 432300,      # miles
    }

    # Bonus: Basic error handling with try except
    try:
        G    = G_dict[(in_units, out_units)]
        Rval = R_dict[(in_units, out_units)]        
    
    except KeyError:
        # expected error from incorrect units
        print("Error: Invalid units: only 'Earth' and 'Sun' are allowed as inputs and only 'kps' and 'mph' are allowed as outputs.")
        print(f"       You requested input units of {in_units} and output units of {out_units}.")
        return

    return (2 * G * m / r) ** 0.5 * Rval

````

````python
# test escape from Earth in different units
# expect 11.186 km/s = 25,020 mph

print(vesc(1,1))
print(vesc(1,1, out_units="mph"))
````

````python
# test escape from the Sun in different units
# expect 618 km/s = 1.4e6 mph

print(vesc(1,1, in_units="Sun"))
print(vesc(1,1, in_units="Sun", out_units="mph"))
````
