## <br><br>The Backend

The **frontend** is the code that you write to create a plot. The **backend** is the code behind the scenes that is translating your code into a visualization. Matplotlib has multiple backends that you can use - some are better at creating **hard** image files (for example, png, jpg, pdf files) and some are better at creating **interactive** visualizations. Some backends are also better for working in Jupyter notebooks or better in scripts.
<br><br>Matplotlib will usually choose an appropriate backend for you. Let's import Matplotlib and see which backend it chose for us.

```python
import matplotlib
matplotlib.get_backend()
```

<br>Matplotlib chose the **inline** backend for me. This is the best backend for visualizing plots in Jupyter notebooks (and Jupyter notebooks are pretty much the best for customizing your plots!). Sometimes your setup might not choose the right backend for your Jupyter notebook and you might notice your plots aren't showing up the way they should. Because of this, you'll often see people including the following code in their Jupyter notebooks:

```python
%matplotlib inline
```

When you see code in a Jupyter notebook that starts with `%`, it is called a **magic function**. It is kind of like a shortcut that lets you do a few special things outside of Python inside the notebook. We're not going to get into magic functions beyond this one. 
<br><br>Jupyter notebooks used to be worse at picking the right backend for Matplotlib, so always using `%matplotlib inline` used to be more necessary. If your notebook didn't automatically select `inline` as your backend when you imported Matplotlib, you should generally use the magic function in your notebooks. If it did automatically choose `inline`, you're probably safe, but if you ever notice things don't seem to be working the way you expect, it might be worth it to run `matplotlib.get_backend()` to double check which backend you're using, and then run `%matplotlib inline` if needed. You only need to run it once, usually near the top of your notebook.

<br>This code will list all the backends available in your notebook:

```python
%matplotlib --list
```

<br>We're not going to cover any other backends today.

## <br><br>Figures and Axes

When creating visualizations in a Jupyter notebook, the rule is:

#### One code cell per plot.

You can't stretch the code for a plot across multiple cells. If you're customizing a lot of details, you may have many many lines of code in that cell.

#### <br>The first line of any plot
There is one line of code that you will always start out with when creating any plot. Antonio calls this "The Fundamental Line":

```python
fig, ax = matplotlib.pyplot.subplots()
```

<br>That's actually the long version of the line. Remember that Matplotlib is a very large package, and that sometimes in Python there are standard nicknames that are used to import packages (pandas is commonly imported as pd, for example). You usually won't actually see people running `import matplotlib` in their notebooks. Instead, we import the `pyplot` library from the `matplotlib` package as `plt`. This is just common. Let's do it here:

```python
import matplotlib.pyplot as plt
```

<br><br>Now our first line of code ("The Fundamental Line") for every plot is:

```python
fig, ax = plt.subplots()
```

All matplotlib plots have one **figure** and one or more **axes**.

An **axes** in matplotlib is **NOT** the same thing as an **axis** on a plot (like the x-axis or y-axis).

The **figure** is the whole visualization, and the **axes** is the individual plot inside the figure. I like to think about it like this

- the *figure* is a wall where you're going to hang your pictures
- the *axes* are the pictures you hang on the wall.

You can have multiple axes in one figure. Here is an example of that:

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

<br><br>We can set some attributes of the **figure** inside the `plt.subplots()` function in this first line:

```python
fig, ax = plt.subplots(nrows=3, figsize=(6,6), facecolor="blue")
```

#### <br><br>Setting attributes with a dictionary

Here's why it's useful to have two ways to customize the look of your plots. You can save a dictionary of your perfect look and reuse it for multiple plots inside the `set()` method function. For example, if you're creating visualizations for a paper to be published in a particular journal, you can save one set of attributes and reuse it on all your plots so that they have a cohesive look. Then, if you need to change one attribute of only one plot, you can change it using the specific `set_` method.

<br>In the dictionary, the keywords from the matplotlib arguments are the keys, and must be stored as strings.

```python
cool_ax = {"facecolor": "lightseagreen", 
           "xticks": (0, 1, 2, 3),
           "xticklabels": ("0", "Year 1", "Year 2", "Year 3")}
cool_fig = {"figsize": (3,1), "facecolor": "darkorange"}
```

<br> When you pass the dictionaries as arguments, you need to **unpack** them by putting `**` in front of the dictionary name:

```python
fig, ax = plt.subplots(**cool_fig)
ax.set(**cool_ax)
```

<br>Now we can use the dictionary, but still change some attributes afterwards:

```python
fig, ax = plt.subplots(**cool_fig)
ax.set(**cool_ax)
ax.set_xticklabels(("0", "Year 2", "Year 4", "Year 6"))
```

### <br><br><br>Selecting an individual axes (ax?)

Sometimes you'll want different features for different **axes** in the same **figure**. For example, they might need different ranges for the x-axis and/or y-axis. As a reminder, here's how we create a figure with multiple axes:

```python
fig, ax = plt.subplots(ncols=2, nrows=2)
plt.show()
```

<br>We can index the `ax` object (like we would any Python object) to select an individual axes. First you index the row, then the column, separated by a comma:

```python
fig, ax = plt.subplots(ncols=2, nrows=2)
ax[0,0].set(facecolor="cyan", xlim=(0,1))
ax[0,1].set(facecolor="magenta", xlim=(0,100))
ax[1,0].set(facecolor="lime", xlim=(0,1000))
ax[1,1].set(facecolor="yellow", xlim=(0,0.1))
plt.show()
```

<br>When indexing an array of axes that only has one column or one row, you only need to index one number:

```python
fig, ax = plt.subplots(ncols=1, nrows=4)
ax[0].set(facecolor="cyan", xlim=(0,1))
ax[1].set(facecolor="magenta", xlim=(0,100))
ax[2].set(facecolor="lime", xlim=(0,1000))
ax[3].set(facecolor="yellow", xlim=(0,0.1))
plt.show()
```

<br>**Near-Earth Object (NEO) Comets:** This dataset contains orbital and physical data for comets that have been classified as near-Earth objects. It was downloaded from the JPL Small-Body Database. Each row represents a comet (or a fragment of a comet). The columns include:

- **per_y**: Orbital period in years
- **first_obs** / **last_obs**: Dates of first and last observation
- **condition_code**: Orbit quality rating (0 = best-determined, 9 = most uncertain)

### <br>Line chart (`ax.plot()`)

Let's start with a line chart that shows the relationship between a comet's semi-major axis (its average distance from the Sun) and its orbital period (how long it takes to complete one orbit). We expect these to be related — comets farther from the Sun should take longer to orbit. We'll sort our data by semi-major axis so the line connects points in order from closest to farthest.

```python
# Drop rows where either semi-major axis or orbital period is missing
kepler = comets.dropna(subset=['a', 'per_y'])
# Keep only shorter-period comets for a cleaner view
kepler = kepler[kepler['per_y'] < 200]
# Sort by semi-major axis so the line goes left to right
kepler = kepler.sort_values('a')
# Show the first five rows
kepler[['full_name', 'a', 'per_y']].head()
```

<br><br>To create a line chart, we use the method `ax.plot()`. We pass it two arguments: data for the x-axis (semi-major axis `a`) and data for the y-axis (orbital period `per_y`). We can also pass additional keyword arguments like line color.

```python
fig, ax = plt.subplots(figsize=(6, 5))
ax.plot(kepler['a'], kepler['per_y'], color="steelblue")
ax.set_title('Orbital Period vs. Distance from the Sun')
ax.set_xlabel('Semi-major axis (AU)')
ax.set_ylabel('Orbital period (years)')
plt.show()
```

<br>The curve sweeps upward — comets farther from the Sun take much longer to orbit. This is Kepler's Third Law: orbital period grows with distance at a rate of T ∝ a^(3/2). The rapid rise at larger distances shows just how dramatically orbital time stretches out.

### <br><br>Scatter plot (`ax.scatter()`)

Let's use a scatter plot to compare the orbital eccentricity and inclination of Jupiter-family and Halley-type comets. Eccentricity measures how elongated the orbit is (0 = circular, near 1 = very elongated). Inclination measures the tilt of the orbit relative to Earth's orbital plane (0° = same plane, >90° = retrograde orbit).

### <br><br>Histogram (`hist()`)

Let's use a histogram to see the distribution of orbital inclinations for Jupiter-family comets and Halley-type comets. For each class, we only need a pandas Series object — a single column — since histograms need only one variable.

```python
# Pandas Series of inclinations for each class
inc_jfc = comets[comets['class'] == 'JFc']['i'].dropna()
inc_htc = comets[comets['class'] == 'HTC']['i'].dropna()
print(inc_jfc.describe())
```

<br>We'll create one figure with two plots:

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].hist(inc_jfc)
ax[0].set_title("Jupiter-family Comets\nInclination Distribution")
ax[1].hist(inc_htc)
ax[1].set_title("Halley-type Comets\nInclination Distribution")
plt.show()
```

<br>The x-axis scales are very different! Jupiter-family comets cluster below 40°, while Halley-type comets span nearly 180°. Let's share the x-axis so we can compare them directly:

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharex=True)
ax[0].hist(inc_jfc)
ax[0].set_title("Jupiter-family Comets\nInclination Distribution")
ax[1].hist(inc_htc)
ax[1].set_title("Halley-type Comets\nInclination Distribution")
plt.show()
```

<br>Now the contrast is clear: Jupiter-family comets orbit near the ecliptic plane (low inclination), while Halley-type comets are spread across all inclinations — many on retrograde orbits above 90°.

The bin sizes, however, are different between the two plots. One has narrow bins and one has wide bins, even though the x-axis is the same. Matplotlib decided the bin sizes for us. We can use code to set exactly where we want the bins to divide. We'll use a range function and set the dividers to every 5° between 0 and 175.

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharex=True)
ax[0].hist(inc_jfc, bins=range(0,175,5))
ax[0].set_title("Jupiter-family Comets\nInclination Distribution")
ax[1].hist(inc_htc, bins=range(0,175,5))
ax[1].set_title("Halley-type Comets\nInclination Distribution")
plt.show()
```

The y-axis displaying the counts is also off between the two histograms. Let's overlay the two histograms on a single axes. As a reminder, to do that we call `ax.hist()` twice.

```python
fig, ax = plt.subplots()
ax.set_title("Orbital Inclination Distribution by Comet Class")
ax.hist(inc_jfc, color="steelblue", label="Jupiter-family (JFc)", bins=range(0,175,5))
ax.hist(inc_htc, color="darkorange", label="Halley-type (HTC)", bins=range(0,175,5))
ax.set_xlabel("Inclination (degrees)")
ax.set_ylabel("Count")
ax.legend()
plt.show()
```

The two classes barely overlap. Jupiter-family comets are thought to originate from the Kuiper Belt — a flat disk — which explains their low inclinations. Halley-type comets originate from the spherical Oort Cloud, which is why their inclinations are scattered across all angles.

### <br><br>Bar graph (`bar()` and `barh()`)

Let's compare how many comets of each class were first discovered in each decade. Automated sky surveys launched around 2000 (LINEAR, NEAT, SOHO, Catalina) dramatically boosted discovery rates. We'll compare Jupiter-family (JFc) and Halley-type (HTC) comets across six decades.

```python
class1 = "JFc"
class2 = "HTC"
# Parse the year from the first_obs date column
comets['first_year'] = pd.to_datetime(comets['first_obs'], errors='coerce').dt.year
# Create one dataframe per decade
df60s = comets[(comets.first_year >= 1960) & (comets.first_year < 1970)]
df70s = comets[(comets.first_year >= 1970) & (comets.first_year < 1980)]
df80s = comets[(comets.first_year >= 1980) & (comets.first_year < 1990)]
df90s = comets[(comets.first_year >= 1990) & (comets.first_year < 2000)]
df00s = comets[(comets.first_year >= 2000) & (comets.first_year < 2010)]
df10s = comets[(comets.first_year >= 2010) & (comets.first_year < 2020)]
# Count comets of each class per decade
class1_counts = [len(df[df['class'] == class1]) for df in [df60s, df70s, df80s, df90s, df00s, df10s]]
class2_counts = [len(df[df['class'] == class2]) for df in [df60s, df70s, df80s, df90s, df00s, df10s]]
print(class1_counts)
```

<br>We also need to create a list for the x-axis — what do the items in our lists represent?

```python
decades = ["1960s", "1970s", "1980s", "1990s", "2000s", "2010s"]
```

<br>We will plot these bar charts side-by-side:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].bar(decades, class1_counts)
ax[0].set_title(class1)
ax[1].bar(decades, class2_counts)
ax[1].set_title(class2)
plt.show()
```

<br>It's hard to compare because they have different ranges for the y-axis. We can match up all the subplots by passing the keyword argument `sharey=True`:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), sharey=True)
ax[0].bar(decades, class1_counts)
ax[0].set_title(class1)
ax[1].bar(decades, class2_counts)
ax[1].set_title(class2)
plt.show()
```

<br>We can improve the look even more by removing the y-axis ticks from the right plot and moving the plots closer together using `fig.tight_layout()`:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), sharey=True)
ax[0].bar(decades, class1_counts)
ax[0].set_title(class1)
ax[1].bar(decades, class2_counts)
ax[1].set_title(class2)
ax[1].yaxis.set_ticks_position('none')
fig.tight_layout()
plt.show()
```
