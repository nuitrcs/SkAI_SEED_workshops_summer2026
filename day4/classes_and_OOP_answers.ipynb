## Exercise 1

```python
import math

class Planet:
    def calculate_escape_velocity(self, mass, radius):
        G = 1.541e-6
        v_esc = math.sqrt((2 * G * mass) / radius)
        # Convert to km/s
        return v_esc * 6371
```

```python
p = Planet()

result = p.calculate_escape_velocity(mass=1, radius=1)
print(result)   # expected: ~11.19

result2 = p.calculate_escape_velocity(mass=317.8, radius=11.2)
print(result2)  # expected: ~59.5  (Jupiter's escape velocity)
```

## Exercise 3

```python
import math

class Planet:
    def __init__(self, name, mass, radius):
        self.name = name
        self.mass = mass
        self.radius = radius
        return

    def calculate_escape_velocity(self):
        G = 1.541e-6
        v_esc = math.sqrt((2 * G * self.mass) / self.radius)
        return v_esc * 6371

    def report_data(self):
        v_esc = self.calculate_escape_velocity()
        print(f"Planet {self.name} has a mass of {self.mass} Earth masses and a radius of {self.radius} Earth radii. Its escape velocity is {v_esc:.2f} km/s.")
        return
```

```python
earth = Planet(name="Earth", mass=1, radius=1)
print(earth.calculate_escape_velocity())  # expected: ~11.19
```

```python
earth.report_data()
# expected output:
# Planet Earth has a mass of 1 Earth masses and a radius of 1 Earth radii. Its escape velocity is 11.19 km/s.
```

```python
# Also test another planet:
jupiter = Planet(name="Jupiter", mass=317.8, radius=11.2)
print(jupiter.calculate_escape_velocity())  # expected: ~59.5
jupiter.report_data()
# expected output:
# Planet Jupiter has a mass of 317.8 Earth masses and a radius of 11.2 Earth radii. Its escape velocity is 59.5X km/s.
```

## Exercise 4

```python
import math

class Planet:
    G = 1.541e-6

    def __init__(self, name, mass, radius):
        self.name = name
        self.mass = mass
        self.radius = radius
        self.moons = []
        return

    def calculate_escape_velocity(self):
        v_esc = math.sqrt((2 * self.G * self.mass) / self.radius)
        return v_esc * 6371

    def report_data(self):
        v_esc = self.calculate_escape_velocity()
        print(f"Planet {self.name} has a mass of {self.mass} Earth masses and a radius of {self.radius} Earth radii. Its escape velocity is {v_esc:.2f} km/s.")
        return

    def add_moon(self, moon_name):
        self.moons.append(moon_name)
        return

    def list_moons(self):
        if len(self.moons) == 0:
            print(f"Planet {self.name} has no moons.")
        else:
            print(f"Moons of {self.name}: {', '.join(self.moons)}")
        return
```

```python
# Check class attribute
print(Planet.G)  # expected: 1.541e-06
```

```python
# Test moon-less planet
earth = Planet(name="Earth", mass=1, radius=1)
earth.list_moons()       # expected: Planet Earth has no moons.
```

```python
# Add moons and list them
jupiter = Planet(name="Jupiter", mass=317.8, radius=11.2)
jupiter.add_moon("Io")
jupiter.add_moon("Europa")
jupiter.add_moon("Ganymede")
jupiter.add_moon("Callisto")
jupiter.list_moons()
# expected: Moons of Jupiter: Io, Europa, Ganymede, Callisto
```

```python
# Confirm Earth's moons list is still empty (no accidental class-level sharing!)
earth.list_moons()  # expected: Planet Earth has no moons.
```

```python
import math

class KeplerSystem:
    def __init__(self, star_name, star_mass):
        self.star_name = star_name
        self.star_mass = star_mass # in Earth masses
        self.orbiting_planets = []
        return

    def add_planet(self, planet, period_days):
        # G in AU^3 * M_Earth^-1 * day^-2
        G = 8.888e-10
        
        total_mass = self.star_mass + planet.mass
        sma = ((G * total_mass * (period_days ** 2)) / (4 * math.pi ** 2)) ** (1/3)
        
        planet.semi_major_axis_au = sma
        self.orbiting_planets.append(planet)
        return

    def get_system_summary(self):
        print(f"Stellar System: {self.star_name}")
        print(f"Star Mass: {self.star_mass} Earth masses")
        print("Orbiting Planets:")
        for planet in self.orbiting_planets:
            v_esc = planet.calculate_escape_velocity()
            sma = planet.semi_major_axis_au
            print(f"- {planet.name}: Mass = {planet.mass} Earth masses, Escape Velocity = {v_esc:.2f} km/s, Semi-major axis = {sma:.2f} AU")
        return
```

```python
# Build the Solar System
# Sun's mass ≈ 333,000 Earth masses
solar_system = KeplerSystem(star_name="Sun", star_mass=333000.0)
earth = Planet(name="Earth", mass=1.0, radius=1.0)
solar_system.add_planet(earth, period_days=365.25)
# Check that semi-major axis was attached to earth
print(earth.semi_major_axis_au)  # expected: ~1.00
```

```python
solar_system.get_system_summary()
# expected output:
# Stellar System: Sun
# Star Mass: 333000.0 Earth masses
# Orbiting Planets:
# - Earth: Mass = 1.0 Earth masses, Escape Velocity = 11.19 km/s, Semi-major axis = 1.00 AU
```

```python
# Let's also add Mars
mars = Planet(name="Mars", mass=0.107, radius=0.532)
solar_system.add_planet(mars, period_days=686.97)
print(mars.semi_major_axis_au)  # expected: ~1.52
solar_system.get_system_summary()
```
